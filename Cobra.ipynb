```python
import escher
import escher.urls
import cobra
import cobra.test
import json
import os
from IPython.display import HTML
import matplotlib.pyplot as plt
import numpy as np
```

# Optimizacion default

```python
model = cobra.io.load_json_model(cobra.test.ecoli_json)
solution=model.optimize()
b = escher.Builder(map_name='iJO1366.Central metabolism',
                   reaction_data=solution.x_dict,
                   # color and size according to the absolute value
                   reaction_styles=['color', 'size', 'abs', 'text'],
                   # change the default colors
                   reaction_scale=[{'type': 'min', 'color': '#cccccc', 'size': 4},
                                   {'type': 'mean', 'color': '#0000dd', 'size': 20},
                                   {'type': 'max', 'color': '#ff0000', 'size': 40}],
                   # only show the primary metabolites
                   hide_secondary_metabolites=True)
b.display_in_notebook()
```

# Eliminar una reaccion importante y analizar el flujo, optimizando para ATP

```python
model = cobra.io.load_json_model(cobra.test.ecoli_json)
model.reactions.get_by_id("ATPM").upper_bound = 1000.
model.change_objective("ATPM")
model.reactions.SUCOAS.delete() #SE ELIMINA SUC0AS
solution=model.optimize()
b = escher.Builder(map_name='iJO1366.Central metabolism',
                   reaction_data=solution.x_dict,
                   # color and size according to the absolute value
                   reaction_styles=['color', 'size', 'abs', 'text'],
                   # change the default colors
                   reaction_scale=[{'type': 'min', 'color': '#cccccc', 'size': 4},
                                   {'type': 'mean', 'color': '#0000dd', 'size': 20},
                                   {'type': 'max', 'color': '#ff0000', 'size': 40}],
                   # only show the primary metabolites
                   hide_secondary_metabolites=True)
b.display_in_notebook()
```

# Ejemplo del notebook eliminando GADP

```python
model = cobra.io.load_json_model(cobra.test.ecoli_json)
model.reactions.GAPD.delete()
b = escher.Builder(map_name='iJO1366.Central metabolism',
                   model=model,
                   # in the map, highlight all reactions that are missing from the model
                   highlight_missing=True)
b.display_in_notebook()
```

Notar que aqui no hay flujo, pues no se uso *reaction_data* parametro del constuctor del mapa solo se muestra el nuevo mapa con la reaccion eliminada remarcada.
